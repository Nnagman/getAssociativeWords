```python
import nltk

tokens_ko = Twitter.morphs(reads)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko, name='사업')

print(len(ko.tokens))        # 전체 토큰 개수
print(len(set(ko.tokens)))   # 고유값 구하기
```

```python
ko.similar('사업')
```

```python
ko.count('방법')
```

```python
#어떤 단어가 문서 내에서 몇번 언급되었는지 확인
plt.figure(figsize=(6,4))
ko.dispersion_plot(['제조업', '서비스업','내용'])
```

```python
# 빈도수 계산 후 상위 100개의 데이터 출력



rank_count=ko.vocab().most_common()#내림차순 정렬

import pandas as pd
title=input("원하는 제목")
Data_df = pd.DataFrame(rank_count)
path="c://python_data/"+title+".csv"
Data_df.to_csv(path,mode='w',encoding='euc-kr')
```
